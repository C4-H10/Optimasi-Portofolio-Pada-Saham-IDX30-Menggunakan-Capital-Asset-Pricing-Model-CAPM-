#PERHITUNGAN BOBOT PORTOFOLIO

### **Penjelasan Variabel**

1.  **RATA RATA**: Merupakan nilai rata-rata pengembalian harian dari masing-masing saham selama periode observasi. Ini merepresentasikan ekspektasi keuntungan (reward) yang diberikan aset tersebut di masa lalu.
2.  **VARIAN**: Mengukur sejauh mana data return tersebar dari nilai rata-ratanya. Dalam investasi, varian digunakan sebagai proksi dari risiko total (total risk) saham tersebut. Semakin besar variannya, semakin tinggi fluktuasi harganya.
3. **BETA**: Indikator risiko sistematis yang mengukur sensitivitas return saham terhadap pergerakan pasar (IHSG).
- $\beta = 1$: Saham bergerak sejalan dengan pasar.
- $\beta > 1$: Saham lebih volatil dari pasar (Agresif).
- $\beta < 1$: Saham kurang volatil dibanding pasar (Defensif).
4.  **ALFA ($\alpha$)**: Komponen imbal hasil yang bersifat independen atau tidak dipengaruhi oleh pergerakan pasar secara keseluruhan.
5.  **E(Ri) CAPM**: Expected Return berdasarkan model CAPM. Dihitung dengan: $$E(R_i) = R_f + \beta_i (E(R_m) - R_f)$$

6. *Excess Return**: Selisih antara return harapan ($E(R_i)$) dengan tingkat bunga bebas risiko ($R_f$). Ini menunjukkan tambahan keuntungan yang diharapkan investor karena memilih aset berisiko dibandingkan menyimpan uang di instrumen yang aman (seperti suku bunga BI).

7. **Bobot (Wi)**: Proporsi alokasi modal pada masing-masing saham dalam portofolio optimal

```python
market_return = return_IHSG.iloc[:, 0]
var_market_sample = market_return.var(ddof=1)
var_market_pop = market_return.var(ddof=0)

rata_rata_suku_bunga = np.mean(suku_bunga_list)
rf = rata_rata_suku_bunga / len(return_saham)
print(f"Rata-rata harian suku bunga BI : {rf}")

betas = {}
for col in return_saham.columns:
    cov_p = np.cov(return_saham[col], market_return, ddof=0)[0][1]
    betas[col] = cov_p / var_market_sample

beta_series = pd.Series(betas)
rata_rata = return_saham.mean()
varian_saham = return_saham.var(ddof=0)
eri = rf + (beta_series * (market_return.mean() - rf))

sigma = return_saham.cov()
inv_sigma = np.linalg.inv(sigma)

excess_return = eri - rf

pembilang = inv_sigma.dot(excess_return)

ones = np.ones(len(eri))
penyebut = ones.T.dot(pembilang)

bobot_wi = pembilang / penyebut

summary_df = pd.DataFrame({
    'RATA RATA': rata_rata,
    'VARIAN': varian_saham,
    'SUM': return_saham.sum(),
    'BETA': beta_series,
    'ALFA': rata_rata - (beta_series * market_return.mean()),
    'E(Ri) CAPM': eri,
    'Excess_Return': excess_return,
    'Bobot (Wi)': bobot_wi
})

summary_df.index.name = 'Saham'
print(f"Total Bobot Wi: {summary_df['Bobot (Wi)'].sum()}")
summary_df
```

```python
print(f"Jumlah saham yang terpilih sebanyak : {len(summary_df)} saham")
print(summary_df[["Bobot (Wi)"]].sort_values(by="Bobot (Wi)", ascending=False))
```

#KINERJA PORTOFOLIO

**Ekspektasi Return Portofolio** : <br>
rata-rata tingkat keuntungan yang diharapkan setiap hari
<br><br>
**Beta Portofolio** : <br>
Jika beta > 1 portofolio lebih volatil dibandingkan pasar.<br>
Jika beta < 1 portofolio lebih stabil. <br>
Jika beta = 1 maka portofolio bergerak mengikuti pergerakan pasar (IHSG).
<br><br>
**Indeks Treynor** :<br>
Jika Indeks Treynor portofolio > Indeks Treynor IHSG, maka portofolio yang disusun sudah optimal


```python
# Rata-rata Return IHSG
rata_rata_ihsg = return_IHSG.mean().iloc[0]

# Suku Bunga Bebas Risiko (Rf)
risk_free = rf

# Beta IHSG
beta_market = 1.0

# Rumus Treynor IHSG
treynor_ihsg = (rata_rata_ihsg - risk_free) / beta_market

# Ekspektasi Return Portofolio
return_portofolio = (summary_df['Bobot (Wi)'] * summary_df['E(Ri) CAPM']).sum()

# Beta Portofolio
beta_portofolio = (summary_df['Bobot (Wi)'] * summary_df['BETA']).sum()

# Rumus Treynor Portofolio
treynor_portofolio = (return_portofolio - risk_free) / beta_portofolio

print("=== KINERJA PORTOFOLIO (INDEKS TREYNOR) ===")
print(f"Return Harian IHSG         : {rata_rata_ihsg:.6f}")
print(f"Return Harian Portofolio   : {return_portofolio:.6f}")
print("-" * 45)
print(f"Beta Portofolio            : {beta_portofolio:.6f}")
print("-" * 45)
print(f"Indeks Treynor IHSG        : {treynor_ihsg:.10f}")
print(f"Indeks Treynor Portofolio  : {treynor_portofolio:.10f}")
print("-" * 45)
```

```python
std_dev = np.sqrt(summary_df['VARIAN'])
summary_df['Sharpe Ratio'] = summary_df['Excess_Return'] / std_dev
summary_df['Treynor Ratio'] = summary_df['Excess_Return'] / summary_df['BETA']
summary_df['Indeks Jensen'] = summary_df['ALFA']
indeks = summary_df[['Sharpe Ratio', 'Treynor Ratio', 'Indeks Jensen']]
indeks
```

# PENGAMBILAN KEPUTUSAN

1. **Indeks Sharpe**: Saham dengan nilai Indeks Sharpe tinggi dapat memberikan imbal hasil yang lebih kompensasif terhadap fluktuasi harganya.

2. **Indeks Treynor**: Portofolio optimal jika Indeks Treynor portofolio lebih besar dari Indeks Treynor harga pasar (IHSG) .

3. **Indeks Jensen**: Saham dengan nilai Indeks Jensen positif maka saham tersebut masuk dalam saham yang sedang mengalami *undervalue*.
